#### Import Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from scipy.stats import pearsonr
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
```

```python
file = "data/game_teams_stats.csv"
raw_df = pd.read_csv(file)
raw_df.head()
```

```python
raw_df.count()
```

```python
drop_na_df = raw_df.dropna(how='any')
drop_na_df.count()
```

```python
drop_na_df.columns
```

```python
data_df = drop_na_df.rename(columns={
    'game_id':'Game ID',
    'team_id':'Team ID',
    'HoA':'Home Or Away',
    'won':'Won',
    'settled_in':'Settled In',
    'head_coach':'Head Coach',
    'goals':'Goals',
    'shots':'Shots',
    'hits':'Hits',
    'pim':'Penalty Minutes',
    'powerPlayOpportunities':'Power Plays',
    'powerPlayGoals':'Power Play Goals',
    'faceOffWinPercentage':'Faceoff Win Percentage',
    'giveaways':'Giveaways',
    'takeaways':'Takeaways'

})
data_df.head()
```

```python
data_df.dtypes
```

```python
data = pd.get_dummies(data_df)
data.head()
```

```python
cor = data.corr()
```

```python
#Using Pearson Correlation
plt.figure(figsize=(20,20))
cor = data_df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
cor_target = abs(cor["Won"])
relevant_features = cor_target[cor_target>0.01]
relevant_features
```

```python
# Highest corrs to Goals - See if they relate

print(data_df[["Power Play Goals","Shots"]].corr())
# print(data_df[["Takeaways","Power Play Goals"]].corr())
```

```python
X = data_df["Power Play Goals"].values.reshape(-1, 1)
y = data_df["Shots"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
plt.scatter(X, y)
```

```python
 # Use sklearn's `train_test_split` to split the data into training and testing

from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
# Create the model

### BEGIN SOLUTION

from sklearn.linear_model import LinearRegression

model = LinearRegression()

### END SOLUTION
```

```python
# Fit the model to the training data. 

### BEGIN SOLUTION

model.fit(X_train, y_train)

### END SOLUTION
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data

from sklearn.metrics import mean_squared_error, r2_score

### BEGIN SOLUTION

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")

### END SOLUTION
```

```python
# Call the `score` method on the model to show the r2 score

### BEGIN SOLUTION
model.score(X_test, y_test)
### END SOLUTION
```
